## Task 1

Calculate the new quantum state after this CNOT operator.

<h3>Solution</h3>

The state before CNOT is $ \sqrttwo \big( a\ket{000} + a \ket{011} + b\ket{100} + b \ket{111} \big) $.

CNOT($q[2]$,$q[1]$) is applied.

If the value of the left most qubit is 1, then the value of the middle qubit is flipped.

Thus, the new quantum state after this CNOT is

$$ \sqrttwo \big( a\ket{000} + a \ket{011} + b\ket{110} + b \ket{101} \big). $$

## Task 2

Calculate the quantum state after after applying this Hadamard operator.

Verify that the new quantum state can be represented as:

$$  
    \frac{1}{2} \ket{00} \big( a\ket{0}+b\ket{1} \big) +
    \frac{1}{2} \ket{01} \big( a\ket{1}+b\ket{0} \big) +
    \frac{1}{2} \ket{10} \big( a\ket{0}-b\ket{1} \big) +
    \frac{1}{2} \ket{11} \big( a\ket{1}-b\ket{0} \big) .
$$

<h3>Solution</h3>

The state before Hadamard is $ \sqrttwo \big( a\ket{000} + a \ket{011} + b\ket{110} + b \ket{101} \big). $

The effect of Hadamard to the left most qubit is given below:

$ H \ket{0yz} \rightarrow \sqrttwo \ket{0yz} + \sqrttwo \ket{1yz} $

$ H \ket{1yz} \rightarrow \sqrttwo \ket{0yz} - \sqrttwo \ket{1yz} $

For each triple $ \ket{xyz} $ in the quantum state, we apply this transformation:

$ 
    \frac{1}{2} \big( a\ket{000} + a\ket{100} \big) + 
    \frac{1}{2} \big( a\ket{011} + a\ket{111} \big) + 
    \frac{1}{2} \big( b\ket{010} - b\ket{110} \big) + 
    \frac{1}{2} \big( b\ket{001} - b\ket{101} \big) .
$

We can rearrange the summation so that we can separate Asja's qubits from the Balvis' qubit:

$ 
    \frac{1}{2} \big( a\ket{000}+b\ket{001} \big) +  
    \frac{1}{2} \big( a\ket{011}+b\ket{010} \big) +  
    \frac{1}{2} \big( a\ket{100} - b\ket{101} \big) + 
    \frac{1}{2} \big( a\ket{111}- b\ket{110} \big)  $.

This is equivalent to

$$  
    \frac{1}{2} \ket{00} \big( a\ket{0}+b\ket{1} \big) +
    \frac{1}{2} \ket{01} \big( a\ket{1}+b\ket{0} \big) +
    \frac{1}{2} \ket{10} \big( a\ket{0}-b\ket{1} \big) +
    \frac{1}{2} \ket{11} \big( a\ket{1}-b\ket{0} \big) .
$$

## Task 3

Asja sends the measurement outcomes to Balvis by using two classical bits: $ x $ and $ y $. 

For each $ (x,y) $ pair, determine the quantum operator(s) that Balvis can apply to obtain $ \ket{v} = a\ket{0}+b\ket{1} $ exactly.

<h3>Solution</h3>

<b>Measurement outcome "00":</b> The state of Balvis' qubit is $ a\ket{0}+b\ket{1} $. 

Balvis does not need to apply any operator.

<b>Measurement outcome "01":</b> The state of Balvis' qubit is $ a\ket{1}+b\ket{0} $. 

If Balvis applies <u>NOT operator</u>, then the state becomes: $ a\ket{0}+b\ket{1} $.

<b>Measurement outcome "10":</b> The state of Balvis' qubit is $ a\ket{0}-b\ket{1} $. 

If Balvis applies <u>Z operator</u>, then the state becomes: $ a\ket{0}+b\ket{1} $.

<b>Measurement outcome "11":</b> The state of Balvis' qubit is $ a\ket{1}-b\ket{0} $. 

If Balvis applies <u>NOT operator</u> and <u>Z operator</u>, then the state becomes: $ a\ket{0}+b\ket{1} $.

## Task 4

Create a quantum circuit with three qubits as described at the beginning of this notebook.

Implement the protocol given above until Asja makes the measurements (included).
- The state of $q[2]$ can be set by the rotation with a randomly picked angle.
- Remark that Balvis does not make any measurement.

At this point, read the state vector of the circuit by using "Simulator.cheat()". 

<i>When a circuit with measurement(s) is simulated by "Simulator.cheat()", the simulator implements the measurement(s) and so returns the state vector accordingly.</i>

Verify that the state of Balvis' qubit is in one of these: $ \ket{v_{00}}$, $ \ket{v_{01}}$, $ \ket{v_{10}}$, and $ \ket{v_{11}}$.

Guess the measurement outcome obtained by "Simulator.cheat()" by checking the entries of the state vector.

<h3>Solution</h3>

```python
# Import all required classes and functions
from projectq import MainEngine
from projectq.ops import H,X,Measure,All, C, NOT, Barrier, Ry
from projectq.backends import CircuitDrawerMatplotlib, Simulator
from projectq.setups.default import get_engine_list
from math import pi, cos, sin
from random import randrange

# Initialize engine objects for simulation and drawings
qdrawer = CircuitDrawerMatplotlib()
qengine = MainEngine(backend = Simulator(), engine_list = [qdrawer]+get_engine_list())

# Define number of qubits and allocate them
qubits = qengine.allocate_qureg(3)
# q2 (qubits[2]) -> Asja's message
# q1 (qubits[1]) -> Asja's qubit in the shared entangled pair
# q0 (qubits[0]) -> Balvis' qubit in the shared entangled pair

print("========================")
print("Quantum message by Asja")
print("========================")
print()

# Rotate the q2 by random angle
r = randrange(100)
theta = 2*pi*(r/100) # radians
print("the picked angle is",r*3.6,"degrees and",theta,"radians")
a = cos(theta)
b = sin(theta)
print("a=",round(a,3),"b=",round(b,3))
print("a*a=",round(a**2,3),"b*b=",round(b**2,3))
print()

# Creating an entanglement between q1 and q0
H | qubits[1]
C(NOT) | (qubits[1], qubits[0])

# define a barrier
Barrier | qubits

# Apply Ry Rotation
Ry(2*theta) | qubits[2]

# CNOT operator by Asja on her qubits where q2 is the controller qubit 
C(NOT) | (qubits[2], qubits[1])

# Hadamard operator by Asja on q2
H | qubits[2]

# define a barrier
Barrier | qubits

# Remember to flush the engine!
qengine.flush()

# Measure Asja's qubits
Measure | qubits[2]
Measure | qubits[1]

# Draw Circuits
qdrawer.draw(qubit_labels={0:'q0',1:'q1',2:'q2'},drawing_order={0:0,1:1,2:2})

print("=======================================")
print("The state vector after the measurements")
print("=======================================")

# Use backend.cheat to print statevector
qstate = qengine.backend.cheat()[1]
print()
print("The quantum state returned by cheat() is")
print(qstate)
print()
print("The state vector is")
for amp in qstate:
    print(round(amp.real,3))
print()

print("==================================")
print("Interpretation of the state vector")
print("==================================")

# we check the nonzero entries in the state vector to determine the measurement outcome
outcomes = ['00','01','10','11']

for i in range(4): # determine which outcome is obtained by tracing the state vector
    if (qstate[2*i].real != 0) or (qstate[2*i+1].real != 0): # nonzero pairs i => (2i,2i+1)
        outcome = outcomes[i] # i-th pair {0,1,2,3}
        print("The measurement outcome is",outcome)
        balvis_state = [ qstate[2*i].real,qstate[2*i+1].real ]
print()

# represent the state in ket notation
qstate_in_ket = "|"+outcome+">"
qstate_in_ket += "("+str(round(balvis_state[0],3))+"|0>+"+str(round(balvis_state[1],3))+"|1>)"
print("The quantum state after the measurements is")
print(qstate_in_ket)
print()


triple = ['000','001','010','011','100','101','110','111']

# represent Balvis' state in ket notation 
# determine the sign of 'a' and 'b' 
balvis_state_in_ket = "|"+outcome+">("
for i in range(len(qstate)):
    amp = qstate[i]
    # compare amp with a,-a,b,-b
    if abs(amp.real-a)<0.000001: 
        balvis_state_in_ket += "+a|"+ triple[i][2]+">"
    elif abs(amp.real+a)<0.000001:
        balvis_state_in_ket += "-a|"+ triple[i][2]+">"
    elif abs(amp.real-b)<0.000001: 
        balvis_state_in_ket += "+b|"+ triple[i][2]+">"
    elif abs(amp.real+b)<0.000001: 
        balvis_state_in_ket += "-b|"+ triple[i][2]+">"
balvis_state_in_ket += ")"        
print("The quantum state of Balvis is")
print(balvis_state_in_ket,", where a=",round(a,3)," and b=",round(b,3))

# Just to prevent exception error prompts. 
# The simulator beckend expects all qubits to be measured/uncomputed
All(Measure) | qubits
```
